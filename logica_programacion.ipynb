##üçì Ejercicio 2: Clasificaci√≥n de calidad de cosecha

**Una cooperativa frut√≠cola necesita clasificar sus lotes de frutas seg√∫n la calidad, en base al peso promedio y al porcentaje de piezas defectuosas. Se requiere crear un programa que, para varios lotes ingresados por el usuario, calcule el peso promedio de una muestra y determine el porcentaje de defectos. En funci√≥n de estos valores, el sistema debe asignar una categor√≠a: ‚ÄúA‚Äù para productos de alta calidad, ‚ÄúB‚Äù para calidad media y ‚ÄúRechazo‚Äù para los que no cumplen los est√°ndares. El programa debe usar funciones para los c√°lculos, bucles para procesar varios lotes y condicionales para establecer la categor√≠a final, mostrando un reporte con los resultados.**

```python
numero_lotes = int(input("Ingrese el n√∫mero de lotes: "))

def calcular_peso_promedio(lista_pesos):
    return sum(lista_pesos) / len(lista_pesos)

def calcular_porcentaje_defectos(lista_defectos):
    return sum(lista_defectos) / len(lista_defectos)

def determinar_categoria(peso_promedio, porcentaje_defectos):
    if peso_promedio > 100 and porcentaje_defectos < 5:
        categoria = "A"
    elif 50 < peso_promedio <= 100 and porcentaje_defectos < 10:
        categoria = "B"
    else:
        categoria = "Rechazo"
    return categoria

pesos = []
defectos = []

for i in range(numero_lotes):
    print(f"\nLote {i+1}:")
    peso = float(input("Ingrese el peso promedio del lote (g): "))
    defecto = float(input("Ingrese el porcentaje de defectos: "))
    pesos.append(peso)
    defectos.append(defecto)

peso_promedio = calcular_peso_promedio(pesos)
porcentaje_defectos = calcular_porcentaje_defectos(defectos)
categoria = determinar_categoria(peso_promedio, porcentaje_defectos)

print("\n--- RESULTADOS ---")
print(f"Peso promedio total: {peso_promedio:.2f} g")
print(f"Porcentaje promedio de defectos: {porcentaje_defectos:.2f}%")
print(f"Categor√≠a final: {categoria}")
```

##üíª Ejercicio 4: Control de inventario de un software agr√≠cola

**Una empresa que desarrolla software para el agro necesita un m√≥dulo que controle el inventario de insumos (fertilizantes, semillas, pesticidas). El programa debe permitir registrar la cantidad inicial de cada producto y simular la salida de unidades seg√∫n las ventas semanales. Con el uso de bucles y condicionales, el sistema debe actualizar los valores y emitir una alerta cuando alg√∫n producto quede por debajo de un nivel m√≠nimo definido como constante. Adem√°s, se debe incluir una funci√≥n que calcule el valor total del inventario restante y muestre un reporte con el stock actualizado.**

```python
NIVEL_MINIMO = 10
PRECIOS = {
    "fertilizante": 5000,
    "semillas": 2000,
    "pesticidas": 3500
}

def registrar_inventario():
    inventario = {}
    print(" Registro inicial de inventario")
    for producto in PRECIOS.keys():
        cantidad = int(input(f"Ingrese la cantidad inicial de {producto}: "))
        inventario[producto] = cantidad
    return inventario

def simular_venta(inventario):
    print("Simulaci√≥n de ventas semanales")
    for producto in inventario.keys():
        venta = int(input(f"Unidades vendidas de {producto}: "))
        inventario[producto] -= venta
        if inventario[producto] < 0:
            inventario[producto] = 0
    return inventario

def verificar_alertas(inventario):
    for producto, cantidad in inventario.items():
        if cantidad < NIVEL_MINIMO:
            print(f"Alerta: {producto} por debajo del nivel m√≠nimo ({cantidad} unidades).")

def calcular_valor_total(inventario, precios):
    total = 0
    for producto, cantidad in inventario.items():
        total += cantidad * precios[producto]
    return total

def mostrar_reporte(inventario, precios):
    print("repote")
    for producto, cantidad in inventario.items():
        valor = cantidad * precios[producto]
        print(f"{producto:<15} | {cantidad:<5} | ${precios[producto]:<15} | ${valor}")
    valor_total = calcular_valor_total(inventario, precios)
    print(f"Valor total del inventario: ${valor_total}\n")

inventario = registrar_inventario()

continuar = "s"
semana = 1

while continuar.lower() == "s":
    print(f" Semana {semana}")
    inventario = simular_venta(inventario)
    verificar_alertas(inventario)
    mostrar_reporte(inventario, PRECIOS)
    continuar = input("¬øRegistrar otra semana?(s/n): ")
    semana += 1

print("\n Programa finalizado. Reporte final:")
mostrar_reporte(inventario, PRECIOS)
```

##‚öôÔ∏è Ejercicio 5: Evaluaci√≥n de desempe√±o de maquinaria

**Una cooperativa agr√≠cola quiere medir el rendimiento de sus tractores en campo. Crea un programa que permita ingresar los datos de varios tractores: horas trabajadas, litros de combustible consumidos y superficie cubierta. El programa debe calcular el rendimiento por hora y el consumo de combustible por hect√°rea. Mediante condicionales, debe clasificar el desempe√±o como ‚ÄúExcelente‚Äù, ‚ÄúAceptable‚Äù o ‚ÄúDeficiente‚Äù, seg√∫n los valores obtenidos. Usa funciones para los c√°lculos, bucles para procesar varios tractores y constantes para los l√≠mites de clasificaci√≥n. Finalmente, muestra un informe con los resultados de cada m√°quina.**

```python
semana = 1
continuar= "s"
while continuar.lower() == "s":
    print("semana")

# Entrada de datos
    horas_trabajadas = input("Ingrese las horas trabajadas por cada tractor: ").split(",")
    litros_combustible = input("Ingrese los litros de combustible consumidos por cada tractor: ").split(",")
    superficie_cubierta = float(input("Ingrese superficie cubierta (hect√°reas): "))

# Funci√≥n para c√°lculos
    def calcular_rendimiento(horas, litros, superficie):
      rendimiento_por_hora = superficie / horas
      consumo_por_hectarea = litros / superficie
      return rendimiento_por_hora, consumo_por_hectarea

# Funci√≥n de clasificaci√≥n
    def clasificar_desempeno(rendimiento_por_hora, consumo_por_hectarea):
      if rendimiento_por_hora >= 10 and consumo_por_hectarea <= 1:
          return "Excelente"
      elif rendimiento_por_hora >= 8 and consumo_por_hectarea <= 2:
          return "Aceptable"
      else:
         return "Deficiente"

# Procesamiento
    resultados = []
    for i in range(len(horas_trabajadas)):
      horas = float(horas_trabajadas[i])
      litros = float(litros_combustible[i])
      rendimiento_por_hora, consumo_por_hectarea = calcular_rendimiento(horas, litros, superficie_cubierta)
      clasificacion = clasificar_desempeno(rendimiento_por_hora, consumo_por_hectarea)
      resultados.append((i + 1, rendimiento_por_hora, consumo_por_hectarea, clasificacion))


    print("\n--- RESULTADOS ---")
    print("Tractor\tRendimiento (ha/h)\tConsumo (L/ha)\tClasificaci√≥n")
    for r in resultados:
      print(f"{r[0]}\t{r[1]:.2f}\t\t\t{r[2]:.2f}\t\t{r[3]}")

    continuar = input("¬øRegistrar otro tractor ?(s/n): ")
    semana += 1
print("programa finalizado.")
```

***

##üöó Ejercicio 3: Optimizaci√≥n de rutas de reparto

Una empresa de mensajer√≠a quiere optimizar sus rutas diarias.
Crea un programa que reciba las distancias (en km) y los tiempos estimados (en minutos) de varios trayectos.
El sistema debe:

    Calcular la velocidad promedio de cada trayecto.

    Determinar cu√°l es la ruta m√°s eficiente (mayor velocidad promedio con menor distancia).

    Calcular el tiempo total del recorrido completo.
    Usa funciones para los c√°lculos, bucles para recorrer las rutas y condicionales para la selecci√≥n de la mejor opci√≥n, mostrando un informe final con las m√©tricas.


```python
def calcular_velocidad(distancia, tiempo):
    return distancia / (tiempo / 60)

# Cantidad de rutas
n = int(input("Ingrese el n√∫mero de rutas: "))

rutas = []
tiempo_total = 0

# Procesar rutas
for i in range(n):
    print(f"\nRuta {i + 1}")
    distancia = float(input("Distancia (km): "))
    tiempo = float(input("Tiempo estimado (minutos): "))

    velocidad = calcular_velocidad(distancia, tiempo)
    rutas.append((i + 1, distancia, tiempo, velocidad))
    tiempo_total += tiempo

# Determinar la ruta m√°s eficiente
ruta_eficiente = rutas[0]
for ruta in rutas[1:]:
    if (ruta[3] > ruta_eficiente[3]) or (ruta[3] == ruta_eficiente[3] and ruta[1] < ruta_eficiente[1]):
        ruta_eficiente = ruta

# Reporte final
print("infomer de rutas")
print("Ruta\tDistancia(km)\tTiempo(min)\tVelocidad(km/h)")
for r in rutas:
    print(f"{r[0]}\t{r[1]:.2f}\t\t{r[2]:.2f}\t\t{r[3]:.2f}")

print(f" Tiempo total del recorrido: {tiempo_total:.2f} minutos")
print(f"Ruta m√°s eficiente: Ruta {ruta_eficiente[0]} con {ruta_eficiente[3]:.2f} km/h y {ruta_eficiente[1]:.2f} km.")
```
